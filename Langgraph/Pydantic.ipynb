Pydantic Basics

```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    name:str
    age:int
    city:str


person = Person(name="Krish" , age = 35, city="Bangalore")
print(person)
```

Optional Fields


```python
from typing import Optional 
class Employee(BaseModel):
    id:int
    name:str
    department:str
    salary:Optional[float] = None   #Optional with default value is None
    is_active: Optional[bool] = True  #Optional with default value as true
```

```python
emp1 = Employee(id=1, name="John", department="IT")
print(emp1)
```

```python
class Classroom(BaseModel):
    classnumber:str
    students:list
    capacity:int
```

```python
value = Classroom(
    classnumber="A101",
    students=["Alice", "John","Veena"],
    capacity=30
)

print(value)
```

Models with Nested Models

```python
class Address(BaseModel):
    street:str
    city:str
    zip:str


class Customer(BaseModel):
    id:str
    name:str
    address:Address


customer = Customer(
    id="3",
    name="Mayukh",
    address={"street": "123 Main Street", "city":"Bangalore", "zip":"543214"}
)

print(customer)
```

Pydantic Fields and customization: It is majorly used to create constraints

```python
from pydantic import BaseModel, Field

#Field helps us set the input contraints


class Item(BaseModel):
    name:str=Field(min_length=2,max_length=50)
    price:float=Field(gt=0, le= 1000)  #
    quantity:int=Field(ge=0)


item=Item(name="Mayukh", price=600, quantity=1)

print(item)
```
